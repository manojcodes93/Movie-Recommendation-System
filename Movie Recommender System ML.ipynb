#### The ast module is used to safely convert string representations of Python objects (like lists or dictionaries) into actual Python objects.

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert(all_movies.iloc[0].genres)
```

```python
all_movies['genres'] = all_movies['genres'].apply(convert)
```

```python
all_movies["keywords"] = all_movies["keywords"].apply(convert)
```

```python
all_movies["cast"][0]
```

```python
import ast

def convert_cast(obj):
    count = 0
    L = []
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
```

```python
all_movies['cast'] = all_movies['cast'].apply(convert_cast)
```

```python
all_movies.crew[0]
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
all_movies["crew"] = all_movies["crew"].apply(fetch_director)
```

```python
all_movies.head()
```

```python
all_movies.overview[0]
```

```python
# Splits each movie overview (string) into a list of words for every row in the 'overview' column
all_movies.overview = all_movies.overview.apply(lambda x:x.split())
```

```python
all_movies.head()
```

### We need to remove the space between every word.. for ex: Science Fiction -> ScienceFiction

```python
all_movies.genres = all_movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
all_movies.keywords = all_movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
all_movies.cast = all_movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
all_movies.crew = all_movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
all_movies['tags'] = all_movies['overview'] + all_movies['genres'] + all_movies['keywords'] + all_movies['cast'] + all_movies['crew']
```

# Vectorization

You should do vectorization here because:

Machines don’t understand text directly — they understand numbers. Vectorization converts text (like movie overviews, cast, genres, etc.) into numerical form.

Makes comparison possible — once text is in vector form, you can measure similarity between movies (e.g., using cosine similarity).

Better search & recommendations — vectorization allows you to recommend movies that are “close” in meaning, not just exact keyword matches.

Efficient computation — operations on vectors (like dot products) are much faster for large datasets than looping through raw text.

Now, let's group the most common keywords from all the movies. Let's say 5000 keywords and see how many words from those 5000 are there in each movie

```python
# CountVectorizer converts text data into numerical vectors
from sklearn.feature_extraction.text import CountVectorizer

# stop_words = 'english' → remove common English words like "the", "is", "and", etc.
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

The Porter Stemmer is used for stemming, which means reducing words to their root/base form.

Ex: 

"loving" → "love"

"played" → "play"

"running" → "run"

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
ps.stem("loving")
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
# compares every movie with every other movie and gives a score:
# 1 = very similar, 0 = not similar.
similarity = cosine_similarity(vectors)
```

```python
# Main function

def recommend(movie):

    # find index of the given movie
    movie_index = new_df[new_df['title'] == movie].index[0]

    # get similarity scores for this movie
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(similarity[movie_index])), reverse = True, key = lambda x:x[1])[1:6]

    # print the top 5 most similar movie titles
    for i in movies_list:
        print(i[0], new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('The Avengers')
```
